```python
# Notebook setup
import xarray as xr
#!pip install netcdf4

import matplotlib.pyplot as plt
```

**NDBC (National Data Buoy Center) Timeseries**

```python
# Open dataset
ds = xr.open_dataset('https://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/44025/44025.ncml')

# Subset the dataset to 1 year
ds = ds.sel(time=slice('2019-01-01','2019-12-31'))
```

```python
ds
```

**Convert Xarray Dataset to Pandas Dataframe**

-- Use Xarray to load NDBC dataset directly from a Thredds server.
-- Xarray is great, especially when dealing with 3D or 4D datasets.
-- Xarray can overcomplicate things. For example, NDBC dataset actually loads with 3 dimensions (time, latitude and longitude), but we only need 1 (time).

-- Here are a few example plotting calls. Can you tell what's different in the output for each?

```python
ds.sea_surface_temperature
```

```python
plt.figure(figsize = (15, 8))

# Built in xarray plotting
ds.sea_surface_temperature.plot();
```

```python
# Plot using matplotlib - This won't work
plt.plot(ds.sea_surface_temperature);
```

```python
plt.figure(figsize = (15, 8))

# Plot using matplotlib - This will, but the units are wrong
plt.plot(ds.sea_surface_temperature.squeeze())
```

```python
plt.figure(figsize = (15, 8))

# Plot using matplotlib - Correctly plotted with time
plt.plot(ds.time,ds.sea_surface_temperature.squeeze());
```

**Convert xarray data to Pandas Dataframe**

To simply things, we can convert our Xarray Dataset to a Pandas Dataframe, which will give use something like a spreadsheet of columns for each variable, and rows for each measurement time.

```python
# Convert to Pandas Dataframe
df = ds.to_dataframe()
df.head(3)
```

```python
df.tail(3)
```

There's still a bit of complexity here because of the multi-dimensional index. If we try to plot this now, we get some crazy labels.

```python
plt.figure(figsize = (15, 8))

df.sea_surface_temperature.plot();
```

Here's how we can properly convert this Dataset to a Dataframe.

```python
# Convert to Pandas Dataframe
df = ds.to_dataframe().reset_index().set_index('time')
df.head(3)
```

```python
plt.figure(figsize = (15, 8))

# Plot using Pandas has built in function
df.sea_surface_temperature.plot();
```

**Customizing Timeseries Plots**

Matplotlib provides a few ways to customize a plot.

**Customizing Lines**

-- linewidth - For example 0.5, 1, 2...  
-- linestyle - For example '-','--', or ':' or other basic or advanced styles  
-- color  
-- marker  
-- label - The name of the line, used in a legend  
-- https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html  
-- https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html  
-- https://matplotlib.org/stable/gallery/index.html

```python
# Line Example
plt.figure(figsize = (15, 7))
plt.plot(df.index, df.sea_surface_temperature, color = 'red', linewidth = 2)
```

```python
# Custom Markers Example
plt.figure(figsize = (15, 7))
plt.plot(df.index, df.sea_surface_temperature, color='red', linestyle='--', marker='s', markerfacecolor='b', 
         markeredgecolor='y', markersize=1)
```

markers = {'.': 'point', ',': 'pixel', 'o': 'circle', 'v': 'triangle_down', '^': 'triangle_up', '<': 'triangle_left', '>': 'triangle_right', '1': 'tri_down', '2': 'tri_up', '3': 'tri_left', '4': 'tri_right', '8': 'octagon', 's': 'square', 'p': 'pentagon', '*': 'star', 'h': 'hexagon1', 'H': 'hexagon2', '+': 'plus', 'x': 'x', 'D': 'diamond', 'd': 'thin_diamond', '|': 'vline', '_': 'hline', 'P': 'plus_filled', 'X': 'x_filled', 0: 'tickleft', 1: 'tickright', 2: 'tickup', 3: 'tickdown', 4: 'caretleft', 5: 'caretright', 6: 'caretup', 7: 'caretdown', 8: 'caretleftbase', 9: 'caretrightbase', 10: 'caretupbase', 11: 'caretdownbase', 'None': 'nothing', 'none': 'nothing', ' ': 'nothing', '': 'nothing'}

**Customizing the Axis**  

-- Axis Title: plt.title('Title')  
-- Legend: plt.legend()  
-- Axes Labels: plt.xlabel('Time') or plt.ylabel('Temperature')  
-- Axes Limits: plt.ylim([-5,5])

```python
df.head(2)
```

```python
# Example of axis customization
plt.figure(figsize = (15, 7))

plt.plot(df.index,df.air_temperature, color='red', label='Air Temperature')
plt.plot(df.index,df.sea_surface_temperature, color='blue', label='Sea Surface Temp')
plt.plot(df.index,df.wave_height, color='green', label='wave_height')

plt.title('Example of axis customization')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.ylim([-15,30])
plt.legend();
```

**Customizing Time Axes Limits**  

-- By default, plots will show the full range of data (x-axis), with a little bit of padding on each side.  
-- To plot the full *Full Time Range* of data: `plt.xlim(df.index.min(), df.index.max())`  
-- To plot a *Specific Time Range*: `plt.xlim(datetime.date(2017,4,1), datetime.date(2019,10,1))` ---> (**import datetime** for this command to run)

```python
import datetime
```

```python
plt.figure(figsize = (15, 7))

plt.plot(df.index,df.air_temperature, color='red', label='Air Temperature')
plt.plot(df.index,df.sea_surface_temperature, color='blue', label='Sea Surface Temp')
plt.plot(df.index,df.wave_height, color='green', label='wave_height')

plt.title('Example of axis customization')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.xlim(datetime.date(2019,4,1), datetime.date(2019,8,1))
plt.ylim([-15,30])
plt.xticks(rotation=45)
plt.legend();
```

**Adding Subplots and Saving**  

-- We can create a figure with multiple plots using the subplots feature.  
-- We can save a figure to a file using fig.savefig()

```python
# Subplot example
fig, (ax1,ax2) = plt.subplots(2,1, sharex=True, figsize=(15,8))

df.air_temperature.plot(ax=ax1)
df.sea_surface_temperature.plot(ax=ax1)
df.wind_spd.plot(ax=ax2, marker='.',linestyle='',markersize=2)

ax1.legend()
ax1.set_ylabel('Temperature (C)')
ax2.set_ylabel('Wind Speed (m/s)')
ax1.set_title('NDBC Station 44025');

# Save the figure to a file
fig.savefig('44025_example.png')
```

```python
# Subplot example
fig, (ax1,ax2,ax3) = plt.subplots(3,1, sharex=True, figsize=(15,12))

df.air_temperature.plot(ax=ax1)
df.sea_surface_temperature.plot(ax=ax1)
df.wind_spd.plot(ax=ax2, marker='.',linestyle='',markersize=2)
df.gust.plot(ax=ax2, marker=',',linestyle='',markersize=3)
df.wave_height.plot(ax=ax3)

ax1.legend()
ax1.set_ylabel('Temperature (C)')
ax1.set_title('NDBC Station 44025');
ax2.legend()
ax2.set_ylabel('Wind Speed (m/s)')
ax3.legend()
ax3.set_ylabel('wave_height')

# Save the figure to a file
fig.savefig('44025_example.png')
```

**Scatterplot**

```python
plt.figure(figsize = (15, 8))

# Method 1 - Using a modified plot() call
plt.plot(df.sea_surface_temperature, df.air_temperature, linestyle='', marker='.', markerfacecolor='y', 
         markeredgecolor='b', markersize=3)

plt.xlabel('Sea Surface Temperature (C)')
plt.ylabel('Air Temperature (C)')
plt.title('NDBC Station 44025 from 1/1/2019 to 1/1/2020');
```

```python
plt.figure(figsize = (15, 8))

# Method 2 - Using scatter()
plt.scatter(df.sea_surface_temperature, df.air_temperature, s=3)

plt.xlabel('Sea Surface Temperature (C)')
plt.ylabel('Air Temperature (C)')
plt.title('NDBC Station 44025 from 1/1/2019 to 1/1/2020');
```
